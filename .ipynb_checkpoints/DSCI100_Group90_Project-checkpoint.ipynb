**Introduction**

In recent years, it has been documented that video game sales have skyrocketed as a result of both increased access to the internet and better living conditions[1]. However, certain video games significantly outperform others in these sales numbers, speculated to be due to the critic and user scores reported on these games. 

To look further into this conjecture, we need to understand the difference between the critic score and the user score of a video game. The critic score of a video game is done by someone who professionally judges the merits of the video game in its ability to target multiple elements of customer retention such as how the game lives up to preview footage, identifying glitches or bugs, the visual aspects of the game, the replayability, the story, etc. These critic reviews are then compiled and averaged into a single critic score. The user score of a video game can be done by anyone who has played the game and compiled across thousands and thousands of user reviews into a single overall score[2].

So the question that we would formulate to delve into this further would be "How does the critic and user score of a video game affect its global sales?"

The dataset used in this report is from Kaggle, containing data for over 13,000 games collected from VGChartz. This dataset was created to try and answer the question “Are video games getting worse?”, which as a result led to the collection of user and critic scores. This dataset was picked to answer our research question as it was one of the only datasets to include both the user and critic scores for each game.


**Methods**

For our project, we will be analyzing the relationship between sales volume and a game’s given user & critic review score. 

The main variables that will be used to make a comparison here are Critic_Score, User_Score, and Total_Shipped.

All the variables are of type double.

Firstly, we wrangled and cleaned our data by removing observations without a critic or a user score. Furthermore, we classified all data points for Total_Shipped that fall 3 standard deviations above the average as an outlier and removed such observations. We do this because any data point that falls that far away from the cluster of points is very clearly not representative of the majority of the datapoints present here. The reason why we use 3 standard deviations specifically is bacause it is the standard used (in UBC STATS as well in multiple data science projects) to determine for boxplots in particular what points would be classified as an outlier. Hence, in our case we will also be using that standard to classify our outliers.

On top of our data set, we provided a mutated column called “Weighted_Score” that provides an average score if we were given both a critic and user score.  Some cases are such that either Critic or User scores are NA, in which case we will have to work with the score that is present so as to include as much data as we can.


```r
### Loading necessary libraries.
library(tidyverse)
library(repr)
library(tidymodels)
library(GGally)
library(ISLR)
library(ggplot2)
options(repr.matrix.max.rows = 6)
```

```r
#reading the data from the raw link from the github repository
videogame <- read_csv(url("https://raw.githubusercontent.com/antaripJKashyap/DSCI100_proj/main/game_sales_data.csv"))
videogame
```

```r
#tidying the data to display predictive capabilities

outlier_limit <- mean(videogame$Total_Shipped) + 3 * sd(videogame$Total_Shipped)
outlier_limit

videogame_combined <- videogame |>
    group_by(Critic_Score, User_Score) |>
    mutate(Weighted_Score = ifelse(!is.na(User_Score) && !is.na(Critic_Score), ((User_Score + Critic_Score)/2), 
        ifelse(is.na(User_Score) && !is.na(Critic_Score), Critic_Score, 
            ifelse(!is.na(User_Score) && is.na(Critic_Score), User_Score, NA)))) |>
    filter(Total_Shipped < outlier_limit) |>
    filter(!is.na(Weighted_Score)) |>
    select(Name, Total_Shipped, Critic_Score, User_Score, Weighted_Score)
videogame_combined
```

Next, we split our dataset into a training and testing set. This is important when evaluating the trustworthiness of our regression model. The training split will be used to train our KNN model and the testing split can then be used to test how accurate our model truly is.

```r
#splitting data into training and testing with the training split having 75% of the data
set.seed(1)
videogame_split <- initial_split(videogame_combined, prop= 0.75, strata= Weighted_Score)
videogame_train <- training(videogame_split)
videogame_test <- testing(videogame_split)
```

We want to first confirm that Weighted score is indeed an appropriate predictor to use. To do so, we will be using ggpairs function to determine its viability. From the ggpairs diagram produced below we can see the appropriate correlations between the predictors.

```r
videogame_eda <- videogame_train |>
      select(-Name) |>
      ggpairs()
videogame_eda
```

```r
videogame_plot <- ggplot(videogame_train, aes(x= Weighted_Score, y= Total_Shipped)) +
geom_point() + labs(x= "Weighted Critic+User Score of the games", y= "Total copies sold")
videogame_plot
summary(videogame_train)
```

From the training data plot-
- We see that sales volume and user & critic review scores are numerical variables 
- The distribution of Total_Shipped points seems to follow a left-skewed pattern with a mean of 0.6038, suggesting an increasing rate of Total Copies Sold as the rating increases. 
- We can see very clearly from this plot that the data does indeed have a positive relationship. However, this relationship does not appear to be linear at all. In fact, we can almost certainly quantify this as an exponential positive relation ship as we see that the gradient of any percieved "line of best fit" will continually increase
- This non-linear relationship suggests it would be appropriate to fit a KNN regression model rather than a linear model

We will now be creating the KNN model using the nearest_neighbor function but first finding the optimal K value.

```r
#training, evaluating and tuning the knn model
set.seed(1)
videogame_spec <- nearest_neighbor(weight_func = "rectangular", neighbors = tune()) |> 
      set_engine("kknn") |>
      set_mode("regression") 

videogame_recipe <- recipe(Total_Shipped ~ Weighted_Score, data = videogame_train) |>
      step_scale(all_predictors()) |>
      step_center(all_predictors())

videogame_vfold <- vfold_cv(videogame_train, v = 5, strata = Weighted_Score)

videogame_workflow <- workflow() |>
  add_recipe(videogame_recipe) |>
  add_model(videogame_spec)

gridvals <- tibble(neighbors = seq(from = 1, to = 300, by = 5))

videogame_results <- videogame_workflow |>
  tune_grid(resamples = videogame_vfold, grid = gridvals) |>
  collect_metrics() 

videogame_min <- videogame_results |>
    filter(.metric == "rmse") |>
    arrange(.metric) |> 
    filter(mean == min(mean))

k_min <- videogame_min |>
    pull(neighbors)
k_min
```

```r
k_val_plot <- videogame_results |>
    ggplot(aes(x= neighbors, y= mean)) +
    geom_point() +
    geom_line() +
    labs(x = "Neighbors", y = "Mean RMSE") +
    ggtitle("RMSE vs K value")
k_val_plot
```

To fit a KNN regression model, we first set up our regression specification and KNN recipe, while stepping and scaling our predictors to ensure it is standardized prior to fitting. We will also tune our regression to find the optimal k value. 

To do so we perform 5-fold cross-validation, form our recipe workflow, create our tuning grid and collect the metrics to find the optimal k value. From our code, we see that k = 281 has the lowest RMSE from a k range of 1-300, with skips of 4. Next, we set up our regression specification and KNN recipe of k = 281.

```r
#we run KNN regression with 281 nearest neighbours
set.seed(1)
videogame_best_spec <- nearest_neighbor(weight_func = "rectangular", neighbors = k_min) |>
         set_engine("kknn") |>
         set_mode("regression")

videogame_best_fit <- workflow() |>
         add_recipe(videogame_recipe) |>
         add_model(videogame_best_spec) |>
         fit(data = videogame_train)
```

```r
#Graph of predicted sales based off of the weighted score (average of critic/user scores) (blue line) 
set.seed(1)
options(repr.plot.width = 10, repr.plot.height = 10)

videogame_preds <- videogame_best_fit |>
          predict(videogame_train) |>
          bind_cols(videogame_train)

videogame_plot <-  
        ggplot(videogame_preds, aes(x = Weighted_Score, y = Total_Shipped)) + 
        geom_point() +
        geom_line(data = videogame_preds, 
            mapping = aes(x = Weighted_Score, y = .pred), 
            color = "blue") +
        xlab("Weighted Critic+User Score") +
        ylab("Total Units Shipped") + 
        ggtitle("Total Units Shipped vs Weighted Critic+User Score") +
        theme(text = element_text(size = 20))
                  
videogame_plot
```

Now that we have generated our KNN model, and also plotted it as per our training data, we now would like to see how the model performs with data it hasn't been trained on.

The main aspect we want to judge for our KNN model is the RMSPE score which is the root mean square percentage error. This will tell us how accurate our KNN model is when we apply it to the real world scenario with data it will not have encountered before.

```r
#performing RSMPE
set.seed(1)
vg_rmspe <- videogame_best_fit |>
        predict(videogame_test) |>
        bind_cols(videogame_test) |>
        metrics(truth = Total_Shipped, estimate = .pred) |>
        filter(.metric == "rmse") |>
        select(.estimate) |>
        pull()
vg_rmspe
```

The above obtained value for our RMSPE is very close to 1. ALthough, we would want the number for the RMSPE to be as low as possible, this number doesnt tell us very much unless perhaps we compare it to our RMSE generated by comparing with our training data

We will also do an RMSE to compare how close the two values are to each other.

```r
set.seed(1)
vg_rmse <- videogame_best_fit |>
        predict(videogame_train) |>
        bind_cols(videogame_train) |>
        metrics(truth = Total_Shipped, estimate = .pred) |>
        filter(.metric == "rmse") |>
        select(.estimate) |>
        pull()
vg_rmse
```

We can see now that the RMSPE and RMSE values are extremely close to each other. This indicates to us that the model is as good a fit for the testing data as it is for the training data. As a result we can be confident that the data trends we see with our training data are likely also present within the testing data as well. 

We can make this assertion because, if for example, those trends did not exist in the testing data, then perhaps either some of the points would have been either way closer or way further than the KNN regression line, thus producing either a far lower or far higher RMSE value respectively.

Therefore, we can conclude that this model is as good a fit we can get for this dataset for all K values below 300.

**Discussion**


As we can see from our KNN fit visualization, we can see a sizable uptick in the total units shipped as the average critic score increases past the 7/10 mark. Prior to this, there is a steady increase in the units shipped up until the score of about 5.5/10. In between the ratings of 5.5 and 7.0, we see a slight dip in the total units shipped which is contrary to the trend in the prior and following ratings.
Past the 9/10 point, we see the trend plateau, with no dips or increases. This could be explained by the very low number of points that are available past this range.

Overall we expected to see a clear positive correlation. In particular, the expectation was to find a large increase in the total units shipped for games with ratings higher than 6.0/10. We expect this because the consumers would perceive a game to be a decently made project once reviewers provide such a rating. This would cause a greater consumer demand for the game causing a rise in the total units shipped to fulfill such a demand.
For the most part, our data follows this trend when applied to our KNN fit. It only disobeys this principle in between the above-mentioned interval.

Such findings could help game developers to decide how many hard copies of their games would be ideal to ship to sellers depending on preliminary critic scores based on the trend produced by our KNN fit model. This would ensure game developers are able to match consumer demand for their project whilst minimizing production costs, and wastage of raw materials caused by the production of excessive hard copies of games.

One obvious question to ask as a result of our analysis would be why the results do not fit the rest of the model’s trends in between the ratings of 5.5 and 7.0.
Further analysis to look into would be what amount of the total demand for a video game project is a result of the critic score and what fraction of the demand is a result of the popularity of the project, the marketing campaign, and the popularity of the developers themselves.
Perhaps the above analysis may explain the sudden change in trends between the 5.5 and 7.0 interval.
